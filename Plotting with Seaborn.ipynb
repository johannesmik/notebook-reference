# Seaborn Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

df = sns.load_dataset('penguins')
df.head()
```

## Swarmplots

Seaborn [Documentation on Swarmplots](http://seaborn.pydata.org/generated/seaborn.swarmplot.html)

```python
ax = sns.swarmplot(x="flipper_length_mm", y="species", hue='island', size=3, data=df)
```

```python
ax = sns.swarmplot(x="flipper_length_mm", y="species", hue='island', dodge=True, size=3, data=df)
```

## Stripplots

[Stripplots](https://seaborn.pydata.org/generated/seaborn.stripplot.html) are similar to swarmplot, but the swarms will have a fixed width.

Also, a _jitter_ can be applied, which makes overlapping points more visible.

```python
ax = sns.stripplot(x="flipper_length_mm", y="species", hue='island', size=5, jitter=True, data=df)
```

## Barplots

[Barplot documentation](https://seaborn.pydata.org/generated/seaborn.barplot.html?)

Normal behavior is to show a confidence interval ("bootstrapping") as error bars.

 - `ci="sd"` shows the standard deviation instead of the confidence interval
 - `ci=None` turns of the error bars completely


```python
ax = sns.barplot(x="species", y="flipper_length_mm", hue="island", errcolor="black", capsize=.1, ci="sd", data=df)
ax.set_ylim(170, 230)
```

Custom estimator can be defined with `estimator`.

```python
ax = sns.barplot(x="species", y="flipper_length_mm", hue="island", estimator=lambda x: x.max(), ci=None, saturation=0.6, data=df)
```
