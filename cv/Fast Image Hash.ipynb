## Fast Image Hash

Described in the article [Google Image Search Explained (medium.com)](https://alibaba-cloud.medium.com/google-image-search-explained-30af8ba9cbea)

This is a fast algorithm to calculate a comparable fingerprint of an image.
n
There are more robust methods, like pHash or SIFT.

### What are similar images?

Similar images are those images, which fingerprints have a Hamming Distance smaller than X (good default is 5).

### Imports


```python
from PIL import Image

def image_hash(image_file):
    im = Image.open(image_file)
    
    # Step 1: Downsize the image to 8x8
    im = im.resize((8, 8), Image.ANTIALIAS)
    display(im)
    
    # Step 2: simplify the color
    im = im.convert("L")
    display(im)
    
    # Step 3: Calculate the average value
    avg = sum(im.getdata()) / 64.
    print("Average:", avg)
    
    # Step 4: Compare the pixel Grayscale to average
    cmp = "".join(list(map(lambda i: "0" if i < avg else "1", im.getdata())))
    print("Comparison:", cmp)
    return cmp
    
def hamming(s1, s2):
    "String based hamming distance"
    h = 0
    for c1, c2 in zip(s1, s2):
        if c1 != c2:
            h += 1
    return h
    
def similar(h1, h2, threshold=5):
    if hamming(h1, h2) < threshold:
        return True
    else:
        return False
```

```python
import os
import IPython.display

cmp1 = image_hash(os.path.expanduser("~/Pictures/foto2.jpg"))
cmp2 = image_hash(os.path.expanduser("~/Pictures/foto2sketch.png"))
print("Hamming Distance:", hamming(cmp1, cmp2))
print("")
print("Similar:", similar(cmp1, cmp2, threshold=5))
```

### Further ideas

 - The 64-bit fingerprint in the above script is a string composed of zeros and ones. This is for demonstrating purpose. It can of course be converted to a single int or written as a hex-number.
